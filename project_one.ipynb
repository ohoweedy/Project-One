```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import datetime as dt

# Import API key
from project_one_api import mlb_api_key
from project_one_api import nfl_api_key
from project_one_api import nhl_api_key

# File to Load
payroll_file = "2015_mlb_payroll.csv"

# Read Purchasing File and store into Pandas data frame
payroll_data = pd.read_csv(payroll_file, encoding="utf-8")

payroll_data['2015 Payroll'] = payroll_data['2015 Payroll'].str.replace(',', '')
payroll_data['2015 Payroll'] = payroll_data['2015 Payroll'].str.replace('$', '')

# Convert columns from string to float
columns_to_convert = ['2015 Payroll']

for column in columns_to_convert:
    payroll_data[column] = payroll_data[column].astype(float)
    
payroll_data.head()
```

```python
# Create MLB wins DataFrame
mlb_pd = pd.DataFrame({"Team": [""], "Win Pct": [""], "Payroll": [""],
                             "Year": [""]})
```

```python
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018]

mlb_all_data = []

# For loop to process and pull all city temperature data
#"https://api.sportradar.us/mlb-t6/seasontd/2014/REG/standings.xml?api_key={your_api_key}"
for year in years:
# try:
    mlb_query_url = f"https://api.sportradar.us/mlb-t6/seasontd/{year}/REG/standings.json?api_key=ucsbd88d6j7cj2hmkq4sdpwe"
    #query_url = f"{url}&appid={api_key}&q={city}&units={units}"

    mlb_standings_response = requests.get(mlb_query_url)
    time.sleep(2)
    mlb_standings = mlb_standings_response.json()
    mlb_all_data.append(mlb_standings)
    
#team_names = []
#win_ratio = []

for z in range(0,len(mlb_standings['league']['season']['leagues'])):
   #loop through division
   for y in range(0,len(mlb_standings['league']['season']['leagues'][z]['divisions'])):
       #loop through each team
       for x in range(0,len(mlb_standings['league']['season']['leagues'][z]['divisions'][y]['teams'])):
           
        team_name = mlb_standings['league']['season']['leagues'][z]['divisions'][y]['teams'][x]['name']
        win_pct = mlb_standings['league']['season']['leagues'][z]['divisions'][y]['teams'][x]['win_p']

        
        # Append into a DataFrame
        mlb_pd = mlb_pd.append({'Team' : team_name, 'Win Pct': win_pct}, ignore_index = True)

        
# except NameError:
#     print("Continuing with the code...")
```
